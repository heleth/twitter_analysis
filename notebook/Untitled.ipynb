```python
import datetime
import os

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
!ls -l ../txt/ | grep tweets
```

```python
filename = [
    'tweets_by_performer_2234.tsv',
][0]
df_org = pd.read_csv(os.path.join('..', 'txt', filename), sep='\t').rename(columns={'Unnamed: 0': 'index'})

df_org.head(20)
```

```python
sns.set(style='whitegrid', font='Meiryo', rc = {'figure.figsize': (10, 15)}, )
ax = sns.barplot(x='cnt', y='performer', data=df_org, palette='Oranges_r')
ax.set_xlabel('ツイート数', fontsize=20)
ax.set_ylabel('歌手名', fontsize=20)
```
